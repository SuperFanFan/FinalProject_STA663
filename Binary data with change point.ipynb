### This corresponds to the first simulation example provided by the paper: detection of change points for binary data. Only a sequence of 0-1 observations are available. The problem formulation is simple but the solution is non-trivial. 

```python
### configuration
from __future__ import division
import os
import sys
import glob
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as stats
import scipy.linalg as la

%matplotlib inline
%precision 4
plt.style.use('ggplot')
```

```python
### Simulation
np.random.seed(1234)
n = 150
m = 2 # number of change points
tau_true = np.array([50, 100])
tau_true_rep = np.array([tau_true[0], tau_true[1] - tau_true[0], n - tau_true[1]])
s_true = np.repeat(range(0, m+1), tau_true_rep)
theta_true = np.array([0.5, 0.9, 0.4])
y = np.random.binomial(1, theta_true[s_true], size = n)
plt.figure(figsize=(8, 6))
plt.plot(np.arange(n) + 1, np.cumsum(y), '-b', linewidth = 2)
plt.xlabel('Time')
plt.ylabel('Cumulative sum')
plt.xlim([0, 151])
plt.title('Binary 0-1 data')
```

```python
######################################
### Estimation for m = 2
######################################
# priors
a, b = 8, 0.1
c, d = 2, 2
# Inits
s = np.zeros(n - m)
for i in range(1, m + 1):
    s = np.append(s, i)

# inits
F_lag, F = np.zeros((n, m + 1)), np.zeros((n, m + 1))                                  
F_lag[0, 0], F[0, 0] = 1, 1
Ptran = np.zeros((m + 1, m + 1))
Ptran[-1, -1] = 1
theta = np.array([0.5, 0.5, 0.5])
nk = np.zeros(m + 1)
vsim = 7000
burn = 1000

# store results
Sn_p = np.zeros((vsim, n))
Theta_p = np.zeros((vsim, m + 1))
# F_lag_sum, F_sum = np.zeros((n, m + 1)), np.zeros((n, m + 1))

for v in range(vsim):
    
    # Update P
    for i in range(m + 1):
        nk[i] = np.sum(s == i)
    nii = nk - 1
    for i in range(m):
        Ptran[i, i] = stats.beta.rvs(a + nii[i], b + 1)
        Ptran[i, i + 1] = 1.0 - Ptran[i, i]
    
    # Update THETA
    for i in range(m + 1):
        uk = np.sum(y[s == i])
        theta[i] = stats.beta.rvs(c + uk, d + nk[i] - uk)
    
    # Update S
    for i in range(1, n):
        for j in range(m + 1):
            F_lag[i,j] = (Ptran[:,j]).dot(F[i - 1,:])
        F[i,:] = F_lag[i,:] * stats.binom.pmf(y[i], 1, theta)
        F[i,:] = F[i,:] / np.sum(F[i,:])
    
    # Sampling s_t
    for i in range(n - 2, 0, -1): # omit update s_n and s_1 because of their degeneracy
        pmfs = F[i,:] * Ptran[:,s[i + 1]]
        pmfs = pmfs / np.sum(pmfs)
        s[i] = np.random.choice(np.arange(m + 1), p = pmfs)
        
    # Save results
    Theta_p[v,] = theta
#     if(v >= burn):
#         F_lag_sum = F_lag_sum + F_lag
#         F_sum = F_sum + F
    # loop control
    if((v + 1) % 200 == 0):
        print v + 1
```
