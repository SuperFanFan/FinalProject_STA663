```python
######################################
### Estimation for m = 1
######################################
n, m = len(y), 1
# priors
a, b = 8., 0.1
c, d = 2., 1.

# Inits - parameters
theta = np.array([2, 2])
s = np.zeros(n - m)
for i in range(1, m + 1):
    s = np.append(s, i)

# Inits - useful quantities
F_lag, F = np.zeros((n, m + 1)), np.zeros((n, m + 1))                                  
F_lag[0, 0], F[0, 0] = 1, 1
Ptran = np.zeros((m + 1, m + 1))
Ptran[0, 0] = 0.9
Ptran[0, 1] = 0.1
Ptran[-1, -1] = 1
nk = np.zeros(m + 1)

# Inits - runs
vsim = 3000
burn = 1000

# Store results
St_p = np.zeros((vsim, n))
Theta_p = np.zeros((vsim, m + 1))
F_lag_sum, F_sum = np.zeros((n, m + 1)), np.zeros((n, m + 1))       

# Gibbs steps
for v in range(vsim):
    
    # Update S_n
    for i in range(1, n):
        for j in range(m + 1):
            F_lag[i,j] = (Ptran[:,j]).dot(F[i - 1,:])
        F[i,:] = F_lag[i,:] * stats.poisson.pmf(y[i], mu = theta)
        F[i,:] = F[i,:] / np.sum(F[i,:])
    
    # Sampling s_t
    for k in range(n - 2, 0, -1): # omit update s_n and s_1 because of their degeneracy
        pmfs = F[k,:] * Ptran[:,s[k + 1]]
        pmfs = pmfs / np.sum(pmfs)
        s[k] = np.random.choice(np.arange(m + 1), p = pmfs)
        
    # Update P
    for i in range(m + 1):
        nk[i] = np.sum(s == i)
    nii = nk - 1
    for i in range(m):
        Ptran[i, i] = stats.beta.rvs(a + nii[i], b + 1)
        Ptran[i, i + 1] = 1.0 - Ptran[i, i]
    
    # Update THETA
    for i in range(m + 1):
        uk = np.sum(y[s == i])
        theta[i] = stats.gamma.rvs(c + uk, scale = 1.0 / (d + nk[i]))
    
    # Save results
    Theta_p[v,:] = theta
    if(v >= burn):
        F_lag_sum = F_lag_sum + F_lag
        F_sum = F_sum + F
    if(v % 200 == 0):
        print v
```
